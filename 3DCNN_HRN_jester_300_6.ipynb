```python
img_depth = 16
train_data = [X_tr_array,label]

(X_train, y_train) = (train_data[0],train_data[1])
print('X_Train shape:', X_train.shape)

train_set = np.zeros((num_samples, img_depth, img_cols,img_rows,3))

for h in range(num_samples):
    train_set[h][:][:][:][:]=X_train[h,:,:,:]
  

patch_size = 16    # img_depth or number of frames used for each video

print(train_set.shape, 'train samples')
```

```python
# CNN Training parameters


nb_classes = 5
```

```python
# convert class vectors to binary class matrices
Y_train = np_utils.to_categorical(y_train, nb_classes)
```

```python

# Pre-processing

train_set = train_set.astype('float32')
print(np.mean(train_set))
train_set -= np.mean(train_set)
print(np.max(train_set))
train_set /=np.max(train_set)
```

```python
import keras
# Define model
#high resolution network
from keras import regularizers
weight_decay = 0.005
l2=keras.regularizers.l2
model = Sequential()
model.add(Conv3D(4,(3,7,7),
                        input_shape=(patch_size, img_cols, img_rows, 3),
                 activation='relu',bias_initializer='ones'))
model.add(MaxPooling3D(pool_size=(1, 2, 2)))
print(model.output_shape)
model.add(Conv3D(8,(3,5,5), activation='relu',bias_initializer='ones'))
model.add(MaxPooling3D(pool_size=(2, 2, 2)))
print(model.output_shape)
model.add(Conv3D(32,(3,5,5), activation='relu',bias_initializer='ones'))
model.add(MaxPooling3D(pool_size=(1, 1, 2)))
print(model.output_shape)
model.add(Conv3D(64,(3,3,5), activation='relu',bias_initializer='ones'))
model.add(MaxPooling3D(pool_size=(1,2, 2)))


#model.add(Reshape((10,4,4,64), name='feature_reshape'))
#model.add(AveragePooling3D(pool_size=(img_depth/2,4,4), strides=(img_depth/2,4,4), 
#                    padding='valid', name='Average_Pooling'))
#model.add(AveragePooling2D(pool_size=(2,2), strides=(2,2), 
#                    padding='valid', name='Average_Pooling'))

model.add(Flatten())
#model.add(GlobalAveragePooling3D())
model.add(Dense(512, activation='relu',bias_initializer='ones'))
model.add(Dropout(0.5))

model.add(Dense(256, activation='relu',bias_initializer='ones'))
model.add(Dropout(0.5))

model.add(Dense(nb_classes,kernel_initializer='normal'))

model.add(Activation('softmax'))
```

```python
import itertools
def confusion_matrix_plot(cm, classes, 
                          title='Normalized Confusion Matrix', 
                          normalize=True, 
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        cm = np.around(cm, decimals=2)
        cm[np.isnan(cm)] = 0.0
    plt.subplots(1, 1, figsize=(8, 8))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```
